```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
draft_year = 2023
base_url = 'https://www.pro-football-reference.com'

players_url = f'{base_url}/years/{draft_year}/draft.htm'
players_html = urlopen(players_url);

soup = BeautifulSoup(players_html, 'html.parser')
table = soup.find('table', id ='drafts')
table_body = table.find('tbody')
rows = table_body.find_all('tr')
```

```python
def get_row_tag(row, data_stat, tag='td'):
    return row.find(tag, attrs={'data-stat': data_stat})

def get_row_tag_text(row, data_stat, tag='td'):
    return get_row_tag(row, data_stat, tag).text

def get_row_link_text(row, data_stat):
    return get_row_tag(row, data_stat).find('a').text
```

```python
df = pd.DataFrame(columns=['player', 'player_url', 'rnd', 'pick', 'tm', 'pos', 'age', 'college', 'height_in', 'weight_lb', 'height_cm', 'weight_kg'])

for row in rows:
    if (row.get('class') and 'thead' in row.get('class')):  
        continue

    player_tag = get_row_tag(row, 'player')
    player_link_tag = player_tag.find('a')

    player = player_link_tag.text
    player_url = f"{base_url}{player_link_tag.get('href')}"

    rnd = get_row_tag_text(row, 'draft_round', tag='th')
    pick = get_row_tag_text(row, 'draft_pick')
    tm = get_row_link_text(row, 'team')
    pos = get_row_tag_text(row, 'pos')
    age = get_row_tag_text(row, 'age')
    college = get_row_link_text(row, 'college_id')

    # Will gather this info later
    height_in = ''
    weight_lb = ''
    height_cm = ''
    weight_kg = ''
    
    player_data = {
        'player': player,
        'player_url': player_url,
        'rnd': rnd,
        'pick': pick,
        'tm': tm,
        'pos': pos,
        'age': age,
        'college': college,
        'height_in': height_in,
        'weight_lb': weight_lb,
        'height_cm': height_cm,
        'weight_kg': weight_kg
    }

    df.loc[len(df.index)] = player_data
    
df.to_csv('players.csv');
df.head(10)
    
```
